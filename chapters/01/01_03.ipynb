## 1.3.8 Parse tree

```{seealso}
* [Parse tree - Wikipedia](https://en.wikipedia.org/wiki/Parse_tree)
```

````python
Touchaku = [None, 0, 3, 4, 10, 11, 12]
kyakusu = len(Touchaku) - 1
for taiken in range(1, 16):
    Kaishi = [0] * (kyakusu + 1)
    Shuryou = [0] * (kyakusu + 1)
    Shuryou[1] = taiken
    for i in range(2, kyakusu + 1):
        Kaishi[i] = max(Touchaku[i], Shuryou[i-1])
        Shuryou[i] = Kaishi[i] + taiken
    saichou = 0
    for i in range(1, kyakusu + 1):
        saichou = max(saichou, Kaishi[i] - Touchaku[i])
    if saichou < 10:
        print('体験時間', taiken, '分間: ', '最長待ち時間', saichou, '分間')
````

````python
Touchaku = [None, 0, 3, 4, 10, 11, 12]
kyakusu = len(Touchaku) - 1
taiken = 1; saichou = 0
while (taiken <= 15) and (saichou < 10):
    Kaishi = [0] * (kyakusu + 1)
    Shuryou = [0] * (kyakusu + 1)
    Shuryou[1] = taiken
    for i in range(2, kyakusu + 1):
        Kaishi[i] = max(Touchaku[i], Shuryou[i-1])
        Shuryou[i] = Kaishi[i] + taiken
    saichou = 0
    for i in range(1, kyakusu + 1):
        saichou = max(saichou, Kaishi[i] - Touchaku[i])
    if saichou < 10:
        print('体験時間', taiken, '分間: ', '最長待ち時間', saichou, '分間')
    taiken = taiken + 1
````

````python
import ast

class IndentingVisitor(ast.NodeVisitor):
    def __init__(self):
        self.indent_level = -1
        self.indent_str = "    " # 4 spaces

    def visit_Assign(self, node):
        print(f"({node.lineno:02d}) {self.indent_level * self.indent_str}", end="")
        print(f"Assign: {ast.unparse(node.targets)} = {ast.unparse(node.value)}")

    def visit_Expr(self, node):
        print(f"({node.lineno:02d}) {self.indent_level * self.indent_str}", end="")
        print(f"Expr: {ast.unparse(node.value)}")

    def visit_For(self, node):
        print(f"({node.lineno:02d}) {self.indent_level * self.indent_str}", end="")
        print(f"For: {ast.unparse(node.target)} in {ast.unparse(node.iter)}")
        self.generic_visit(node)
        
    def visit_While(self, node):
        print(f"({node.lineno:02d}) {self.indent_level * self.indent_str}", end="")
        print(f"While: {ast.unparse(node.test)}")
        self.generic_visit(node)

    def visit_If(self, node):
        print(f"({node.lineno:02d}) {self.indent_level * self.indent_str}", end="")
        print(f"If: {ast.unparse(node.test)}")
        self.generic_visit(node)

    def generic_visit(self, node):
        if hasattr(node, 'body') and isinstance(node.body, list):
            self.indent_level += 1 # Increase indent
            for item in node.body:
                self.visit(item)
            self.indent_level -= 1 # Decrease indent (dedent)
        else:
            super().generic_visit(node)


tree = ast.parse(code)
visitor = IndentingVisitor()
visitor.visit(tree)
````
