## Thales（タレス）とは

### 生涯
- **生**: 紀元前 624 年（おおよそ）
- **死**: 紀元前 546 年
- **出身**: ミレトス（現在のトルコ）
- **時代**: 古代ギリシャの哲学者・数学者

### 業績
Thales は以下の発見で知られています：
1. **幾何学の父** — 幾何学を「体系的な学問」として確立した最初の人物
2. **Thales の定理** — 「円の直径に対する円周角は直角である」
3. **電気磁気** — 琥珀（amber）を磨くと異なる物質を引き付けることを発見
4. **哲学** — 「万物は水からできている」という初期 monism（唯物論）を唱えた

## 古代ギリシャにおける幾何学

### なぜ幾何学なのか？

古代ギリシャでは、数学は以下の目的で発展しました：

**1. 実用的需要**
- 土地測定（エジプトのナイル川の洪水後の土地区画）
- 建築（神殿、劇場）
- 天文学（星座、季節の予測）

**2. 哲学的興味**
- 「世界の根本原理は何か？」
- 「なぜ自然界はこのような形をしているのか？」
- 「真理とは何か？」

**3. 美的関心**
- 円、正多角形の完全性
- 対称性と調和

## Thales の定理の発見

### 問題設定

Thales は以下の問いに直面しました：

> **「円とは何か？」**
>
> 円は「中心から等距離にあるすべての点の軌跡」として定義される。
>
> では、円の上にある 3 つの点と円の中心との間に、特別な角度関係があるだろうか？

### 特別な場合: 直径

Thales が特に注目したのは、**円の直径の端点** ($A$, $B$) と、円周上の任意の点 ($P$) によって形成される角 $\angle APB$ でした。

$$\text{Thales の観察}: \angle APB = 90° \text{ (常に)}$$

### なぜこれが重要か？

1. **普遍的性質** — すべての円、すべての点に対して成立
2. **作図への応用** — コンパスと定規のみで直角を作図できる
3. **建築への応用** — 神殿や劇場の隅角を正確に作図
4. **証明の例** — 幾何学が「実験的観察」ではなく「論証的体系」であることを示す
