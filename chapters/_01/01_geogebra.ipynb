# 01_geogebra.ipynb: GeoGebra での動的構築

## 学習目標
- GeoGebra で円と直径を構築する
- 円周上の点を動かしながら、角度の変化を観察する
- Thales の定理の「動的な」検証を体験する

## GeoGebra とは

**GeoGebra** は、インタラクティブな幾何学ソフトウェアです。

### 特徴
- **動的作図** — 点や線をドラッグして動かすと、他の要素が自動で更新される
- **測定** — 角度、距離、面積などを自動的に計算・表示
- **プログラミング** — JavaScript API により、自動化や外部制御が可能

### Thales の定理との関係
- 黒板での実験では「測定誤差」が避けられない
- GeoGebra では、点を動かしながら「常に 90 度」が保たれることを確認できる
- これが「数学的真実」と「物理的現実」の違いを理解する助けになる

## GeoGebra でのステップバイステップ構築

以下の操作を、GeoGebra のオンラインエディタで実行してください：

**Web**: https://www.geogebra.org/geometry

### ステップ 1: 中心と 1 点を設置

1. ツールバーから「点」を選択
2. キャンバス上に 2 つの点を配置：
   - 点 $O$ （中心）
   - 点 $A$ （円周上の点）

**コマンド形式** (JavaScript/GeoGebra API):
```javascript
api.evalCommand("O = (0, 0)");      // 中心
api.evalCommand("A = (3, 0)");      // 円周上の点（x = 3）
```

### ステップ 2: 円を描く

1. ツールバーから「圓（中心と点）」を選択
2. 中心 $O$ を選択、次に点 $A$ を選択
3. 半径 $|OA|$ の円が描かれる

**コマンド形式**:
```javascript
api.evalCommand("c = Circle(O, A)");  // 中心 O、点 A を通る円
```

**結果**: 円 $c$ が定義される。中心は $O = (0, 0)$、半径は $|OA| = 3$

### ステップ 3: 直径を定義

1. $A$ の対称点 $B$ を作成（直径の反対側）
2. コマンド: `B = Point(c)` （または）`B = Reflect(A, O)` （中心に対する反射）
3. または、手動で：ツール「直線」を使って $A$ と $O$ を通る直線を引き、その交点として $B$ を取得

**コマンド形式**:
```javascript
api.evalCommand("B = Reflect(A, O)");  // O に対する A の反射 → B は直径の反対側
```

**結果**: 点 $A$ と $B$ は直径の両端、$O$ は中心（直径の中点）

### ステップ 4: 円周上の移動可能な点を配置

1. 「円周上の点」を選択
2. 円 $c$ 上をクリック
3. これが点 $P$ になる

**コマンド形式**:
```javascript
api.evalCommand("P = Point(c)");  // 円周上の点
```

**重要**: $P$ は **自由に動かせる** 点（スライダーと同じ）

### ステップ 5: 線分と角度を描く

1. ツール「線分」を使って、$P$ と $A$、$P$ と $B$ を結ぶ
2. ツール「角度」を使って、$\angle APB$ を測定

**コマンド形式**:
```javascript
api.evalCommand("seg1 = Segment(P, A)");   // 線分 PA
api.evalCommand("seg2 = Segment(P, B)");   // 線分 PB
api.evalCommand("angle = Angle(A, P, B)"); // 角度 ∠APB
```

**結果**: 画面に $\angle APB$ の度数が表示される

### ステップ 6: 動的検証

**重要なポイント**: 点 $P$ をマウスで **ドラッグして移動** してください

1. $P$ をドラッグする
2. $P$ の位置が変わるたび、角度 $\angle APB$ は自動的に再計算される
3. **観察**: どんな位置に $P$ を動かしても、$\angle APB = 90°$ である

この現象が **Thales の定理** の本質です。

## 完成図

```
       P (円周上、移動可能)
      / \
     /   \
    /     \
   A-------O-------B  (直径)
    \     /
     \   /
      \ /
   (円 c がこれらを通る)

∠APB = 90° (常に)
```

## 発展: 複数の点での検証

### 方法 1: スライダーで自動化
1. ツール「スライダー」を使って、パラメータ $t$ を定義（0 から 360 まで）
2. 点 $P$ の位置を $t$ に依存させる：
   ```javascript
   P = (3*cos(t°), 3*sin(t°))  // 中心 O = (0,0)、半径 3
   ```
3. $t$ を動かすと、$P$ が円を一周する
4. 常に $\angle APB = 90°$ である

### 方法 2: 複数の点を配置
1. 「円周上の点」ツールで、複数の点 $P_1, P_2, P_3, \ldots$ を配置
2. 各点に対して角度を測定
3. すべての角度が 90° であることを確認

## 黒板との比較

| 項目 | 黒板描画 | GeoGebra |
|------|--------|----------|
| **精度** | 測定誤差が避けられない | 完全に 90° |
| **動的性** | 一度描いたら固定 | ドラッグで動的に変化 |
| **説得力** | 「だいたい 90° 程度」 | 「常に 90° である」証拠 |
| **学習効果** | 「なぜずれるのか」を考察 | 「数学的な定理」を実感 |

両者は補完的です。黒板での「なぜずれるのか」という問いから、GeoGebra での「いつも 90° である」という確認へ、段階的に進みます。
