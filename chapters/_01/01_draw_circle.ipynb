## 円とは何か？

### 定義
**円** (circle) とは、平面上の固定された 1 点（中心）から、一定の距離（半径）にあるすべての点の集合である。

$$\text{Circle} = \{P \in \mathbb{R}^2 : |P - O| = r\}$$

ここで、$O$ は中心、$r$ は半径。

### 直感的理解
- 中心 $O$ から等距離（半径 $r$）にある点を、くるっと一周つなぐ
- その軌跡が「円」

### コンパスでの実装
コンパスの 2 本の足の距離を「半径」に設定し、一方の足を中心に固定して、もう一方の足で描く

## 「完全な円を描く」ことの困難性

### 実際の問題
1. **機械的誤差** — コンパスの足がブレることで、半径が変動
2. **人間の動き** — 手がグラグラして、円が歪む
3. **回転の開始・終了** — スタート地点に完全に戻すことは難しい
4. **表面の不均一** — 黒板の表面が平坦でない

### 数学的な問題
- **実際には「無限多角形」** — 円を描くのに有限の時間がかかるため、実際には多角形に近い
- **離散 vs 連続** — 現実の描画は離散的だが、数学的な円は連続的

### 結論
**数学における「円」は、物理的な描画とは異なる、理想化された概念である。**

## 直径と円周角の視覚化

### ステップ 3: 角度を測定

## 理由の分析

### 1. 測定誤差
- 分度器の読み取り精度：±1 度程度
- 線の太さによる読み取り位置のズレ

### 2. 描画誤差
- コンパスで描いた円が完全ではない
- 直線が曲がっている、または交点がズレている

### 3. 幾何学的困難
- 現実の描画には、常に「微小な誤差」が伴う
- 数学的な証明とは異なり、測定は「近似」である

### 結論
