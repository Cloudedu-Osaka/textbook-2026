## Thales の定理（再掲）

**定理**: 円の直径に対する円周角は直角である。

$$\text{Circle with diameter } AB \implies \angle APB = 90° \text{ for any point } P \text{ on the circle (except } A, B)$$

## 証明 1: 幾何学的証明

### 準備
- 円の中心を $O$ とする
- 直径の両端を $A$, $B$ とする（つまり $AB$ は直径）
- 円周上の点を $P$ とする（$P \neq A, B$）

### 証明

**Step 1: 三角形 $OPA$ と $OPB$ は二等辺三角形**

円の定義より、すべての円周上の点は中心から等距離（半径）にある。

$$|OA| = |OP| = |OB| = r \text{ (半径)}$$

したがって:
- 三角形 $OPA$: $|OA| = |OP|$ → 二等辺三角形
- 三角形 $OPB$: $|OP| = |OB|$ → 二等辺三角形

**Step 2: 底角が等しい**

二等辺三角形では、等しい辺に対する底角は等しい。

三角形 $OPA$ で:
$$\angle OPA = \angle OAP \quad (= \alpha \text{ とする})$$

三角形 $OPB$ で:
$$\angle OPB = \angle OBP \quad (= \beta \text{ とする})$$

**Step 3: 角度の和**

点 $P$ での角度は：
$$\angle APB = \angle OPA + \angle OPB = \alpha + \beta$$

**Step 4: 三角形 $APB$ の内角の和**

三角形 $APB$ の内角の和は $180°$：
$$\angle PAB + \angle ABP + \angle APB = 180°$$

ここで：
- $\angle PAB = \angle OAP = \alpha$ （直線 $AB$ 上での角度）
- $\angle ABP = \angle OBP = \beta$ （直線 $AB$ 上での角度）

したがって：
$$\alpha + \beta + \angle APB = 180°$$

また、$\angle APB = \alpha + \beta$ なので：
$$(\alpha + \beta) + (\alpha + \beta) = 180°$$
$$2(\alpha + \beta) = 180°$$
$$\alpha + \beta = 90°$$

**結論**:
$$\boxed{\angle APB = 90°}$$

$\blacksquare$

## 証明 2: ベクトル的証明（内積を使用）

### 準備

座標系を設定:
- 中心 $O$ を原点 $(0, 0)$
- 直径 $AB$ を $x$ 軸上に配置
- $A = (-r, 0)$, $B = (r, 0)$ （$r$ は半径）

円周上の点 $P = (x, y)$ は円の方程式を満たす:
$$x^2 + y^2 = r^2$$

### 証明

**Step 1: ベクトルの定義**

$$\vec{PA} = A - P = (-r - x, -y)$$
$$\vec{PB} = B - P = (r - x, -y)$$

**Step 2: 内積を計算**

$$\vec{PA} \cdot \vec{PB} = (-r - x)(r - x) + (-y)(-y)$$
$$= -(r + x)(r - x) + y^2$$
$$= -(r^2 - x^2) + y^2$$
$$= -r^2 + x^2 + y^2$$

**Step 3: 円の方程式を適用**

$P$ は円周上にあるので $x^2 + y^2 = r^2$:

$$\vec{PA} \cdot \vec{PB} = -r^2 + r^2 = 0$$

**Step 4: 直交の定義**

2 つのベクトルの内積が 0 であれば、それらは直交（垂直）している:

$$\vec{PA} \perp \vec{PB} \iff \angle APB = 90°$$

**結論**:
$$\boxed{\angle APB = 90°}$$

$\blacksquare$

## 証明 3: 三角関数的証明

### 準備

中心を原点とする円上の点をパラメータ化:
$$P = (r\cos\theta, r\sin\theta)$$

直径の両端:
$$A = (-r, 0), \quad B = (r, 0)$$

### 証明

**Step 1: 角度の公式（Law of Cosines）を使う**

三角形 $APB$ における辺の長さ:
$$|AP|^2 = (r\cos\theta - (-r))^2 + (r\sin\theta - 0)^2$$
$$= (r\cos\theta + r)^2 + r^2\sin^2\theta$$
$$= r^2(\cos\theta + 1)^2 + r^2\sin^2\theta$$
$$= r^2[(\cos^2\theta + 2\cos\theta + 1) + \sin^2\theta]$$
$$= r^2[1 + 2\cos\theta + 1]$$
$$= r^2[2 + 2\cos\theta]$$
$$= 2r^2(1 + \cos\theta)$$

同様に:
$$|BP|^2 = (r\cos\theta - r)^2 + (r\sin\theta)^2$$
$$= r^2[(\cos\theta - 1)^2 + \sin^2\theta]$$
$$= r^2[\cos^2\theta - 2\cos\theta + 1 + \sin^2\theta]$$
$$= r^2[2 - 2\cos\theta]$$
$$= 2r^2(1 - \cos\theta)$$

直径:
$$|AB|^2 = (2r)^2 = 4r^2$$

**Step 2: Cosine Rule（余弦定理）を適用**

$$|AB|^2 = |AP|^2 + |BP|^2 - 2|AP||BP|\cos(\angle APB)$$

$$4r^2 = 2r^2(1 + \cos\theta) + 2r^2(1 - \cos\theta) - 2|AP||BP|\cos(\angle APB)$$

$$4r^2 = 2r^2 + 2r^2\cos\theta + 2r^2 - 2r^2\cos\theta - 2|AP||BP|\cos(\angle APB)$$

$$4r^2 = 4r^2 - 2|AP||BP|\cos(\angle APB)$$

$$0 = -2|AP||BP|\cos(\angle APB)$$

**Step 3: 結論**

$|AP| > 0$ および $|BP| > 0$ なので：
$$\cos(\angle APB) = 0$$
$$\angle APB = 90°$$

**結論**:
$$\boxed{\angle APB = 90°}$$

$\blacksquare$

## Python での数値検証

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle as CircleArtist

# 設定
r = 1.0  # 半径
angles = np.linspace(0, 2*np.pi, 13)[1:-1]  # 0と2πを除いた12個の角度

# 結果を格納
results = []

for theta in angles:
    # 円周上の点
    P = np.array([r * np.cos(theta), r * np.sin(theta)])
    A = np.array([-r, 0])
    B = np.array([r, 0])
    
    # ベクトル
    PA = A - P
    PB = B - P
    
    # 内積
    dot_product = np.dot(PA, PB)
    
    # 角度を計算（度数法）
    cos_angle = dot_product / (np.linalg.norm(PA) * np.linalg.norm(PB))
    cos_angle = np.clip(cos_angle, -1, 1)  # 数値誤差対策
    angle_deg = np.degrees(np.arccos(cos_angle))
    
    results.append({
        'theta': theta,
        'P': P,
        'angle_deg': angle_deg,
        'dot_product': dot_product
    })

# 結果の表示
print("\n" + "="*70)
print("数値検証: 異なる P の位置での ∠APB")
print("="*70)
print(f"{'Position':15s} {'P (x, y)':30s} {'∠APB':15s} {'Dot Prod':15s}")
print("-"*70)

for i, result in enumerate(results):
    theta_deg = np.degrees(result['theta'])
    px, py = result['P']
    print(f"{theta_deg:6.1f}°{'':<8s} ({px:6.3f}, {py:6.3f})   "
          f"{result['angle_deg']:6.2f}°     {result['dot_product']:10.6f}")

print("\n" + "="*70)
print(f"結果: すべての角度が 90°（±誤差範囲内）")
print(f"Pass Rate: {sum(1 for r in results if abs(r['angle_deg'] - 90) < 0.1) }/{len(results)}")
print("="*70)
```

## グラフィカル検証

```python
# グラフの作成
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

# 軸1: 円と複数の角度を示す
ax = axes[0, 0]
circle = CircleArtist((0, 0), r, fill=False, edgecolor='blue', linewidth=2)
ax.add_patch(circle)

A = np.array([-r, 0])
B = np.array([r, 0])

# 直径を描画
ax.plot([A[0], B[0]], [A[1], B[1]], 'k-', linewidth=2, label='Diameter AB')

# 複数の P での三角形を描画
for i, theta in enumerate(angles[::2]):  # 2つおきにプロット
    P = np.array([r * np.cos(theta), r * np.sin(theta)])
    ax.plot([A[0], P[0], B[0]], [A[1], P[1], B[1]], 'r--', alpha=0.3)
    ax.plot(P[0], P[1], 'ro', markersize=5)

# 点の標識
ax.plot(A[0], A[1], 'bs', markersize=10, label='A')
ax.plot(B[0], B[1], 'gs', markersize=10, label='B')
ax.plot(0, 0, 'k+', markersize=15, markeredgewidth=2, label='Center O')

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Thales の定理: 複数の P での確認', fontsize=12, fontweight='bold')

# 軸2: 角度のグラフ
ax = axes[0, 1]
thetas_deg = [np.degrees(r['theta']) for r in results]
angles_deg = [r['angle_deg'] for r in results]
ax.plot(thetas_deg, angles_deg, 'b-', linewidth=2, marker='o', label='∠APB')
ax.axhline(y=90, color='r', linestyle='--', linewidth=2, label='Expected: 90°')
ax.fill_between(thetas_deg, 89.9, 90.1, alpha=0.2, color='green', label='Error range')
ax.set_xlabel('P の位置 (θ [度])', fontsize=11)
ax.set_ylabel('∠APB [度]', fontsize=11)
ax.set_title('角度の変化: すべての位置で 90°', fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_ylim(85, 95)

# 軸3: 内積のグラフ
ax = axes[1, 0]
dot_products = [r['dot_product'] for r in results]
ax.plot(thetas_deg, dot_products, 'g-', linewidth=2, marker='s')
ax.axhline(y=0, color='r', linestyle='--', linewidth=2, label='Expected: 0')
ax.fill_between(thetas_deg, -0.01, 0.01, alpha=0.2, color='red', label='Error range')
ax.set_xlabel('P の位置 (θ [度])', fontsize=11)
ax.set_ylabel('PA · PB（内積）', fontsize=11)
ax.set_title('内積: すべての位置で 0（直交）', fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend()

# 軸4: 証明方法の比較
ax = axes[1, 1]
ax.axis('off')

proof_text = """
3 つの証明方法の比較

幾何学的証明:
  • 二等辺三角形の性質を利用
  • 古典的なユークリッド幾何学
  • 直感的に理解しやすい

ベクトル的証明:
  • ドット積が 0 ⇒ 直交
  • 代数的で簡潔
  • 高次元への一般化が容易

三角関数的証明:
  • 余弦定理を活用
  • パラメータ化による証明
  • 解析学的なアプローチ

結論: すべて同じ真実に到達
"""

ax.text(0.1, 0.5, proof_text, fontsize=11, family='monospace',
        verticalalignment='center', bbox=dict(boxstyle='round', 
        facecolor='wheat', alpha=0.5))

plt.suptitle('Thales の定理: 包括的検証', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\nグラフィカル検証を表示しました")
```

## 応用 1: Thales の定理の逆

```python
print("""
逆定理 (Converse of Thales' Theorem):

「∠APB = 90° ならば、AB は円の直径である」

この逆も真です。つまり、Thales の定理は双方向に成立します:

  AB は直径 ⟺ ∠APB = 90° (for all P on the circle)

これは、直角三角形の外接円の中心が、斜辺の中点であることを意味します。
""")
```

## 応用 2: 3 次元への拡張

```python
print("""
3 次元への拡張:

2 次元の定理を 3 次元へ拡張すると:

「球の直径に対する外接角は直角である」

- 円 → 球
- 直径 → 直径（球の中心を通る弦）
- 円周上の点 → 球面上の点
- 円周角 → 外接角

ベクトル的証明は 3 次元でもそのまま適用できます:

  PA · PB = 0 ⟹ ∠APB = 90°

より一般的には、任意の n 次元ユークリッド空間でも成立します。
""")
```
