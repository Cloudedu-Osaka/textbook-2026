# 第 0 回: この教材はどう作られているか — AI と人間の役割分担

> 本書を読み始める前に、この教材がどういう過程で作られているかを共有します。AI に対して **拒絶感** を持つ人も、AI を **信頼しすぎる** 人も、両方の立場の人に同じ事実を伝えることが目的です。

---

## 1. この教材の中核は AI の生成物ではない

本講義の縦軸 — バビロニアの恒等式 $(x+y)^2 = (x-y)^2 + 4xy$ から始まり、Menaechmus・Apollonius・Kepler・Newton・Dandelin・Feynman を貫いて、最後に五心と楕円二焦点の対称性に至る四千年の系譜 — は、Higashida が **五年かけて発見・整理した** ものです。

これは AI に書かせて生まれたものではありません。むしろ、五年間の継続的な学習・対話・読書・授業実践の積み重ねの中で **教師の頭の中に少しずつ結晶した** ものです。LLM（Claude や GPT などの AI）は、定義上、セッションを跨いで記憶を持ちません。五年間の継続的学習という時間構造を、AI は内在的に持っていない。

**ただし**、これは越えられない壁ではありません。**Smalltalk は記憶を持つ言語環境でした**。プログラマが書いたコード、実行中のオブジェクト、デバッガ・ブラウザの状態、その全てが image としてセッションを跨いで保持される — 「個人の認知容量を超える知識を、システム側に持続させる」問題を、四十年前に Smalltalk は実装として解いていました（詳細は §4）。

本講義の教材作成・システム開発・授業実践は、この image vision に **AI の長期記憶を持たせる装置** として homage する取り組みの一環です。五年間の発見・対話・判断を Markdown / Jupyter / SPEC として外部化し、LanceDB と MCP を介して AI が前回までの文脈を session ごとに **再ロード** できる経路を整える — これにより、AI 単独では持てない「時間を跨ぐ知識の連続性」が image 経由で回復します。

つまり、本講義の主張は **人間にしか辿れなかった経路を経て獲得され**、かつ **AI が image を介して継続的に参照できる形式に整えられた** ものです。Smalltalk が半世紀前に解いた「個人の時間構造の外部化」を、AI 時代に組み直す試みでもあります。

## 2. では AI は何をしているか — 軸沿いに routing する道具として

AI に依頼しているのは、次のような作業です:

- 教師の五年間の発見を **構造化された軸として整理し直す**（image と呼びます。これは Smalltalk の image vision の現代再演で、半世紀前から確立された設計です）
- 学生がアクセスしやすい形に **整形** する（例: 数式を Python コードに翻訳する、図を作る、定理の証明を順序立てて並べる）
- 教師の判断を **複数の表現で言い換える**（例: 同じ重心の話を、Archimedes の天秤・面積分割・バリセントリック座標の三通りで書く）

重要なのは: AI は **軸を発明していない**。軸は教師が五年かけて見つけたもので、AI はそれに沿って具体化作業を担っている。例えるなら、教師が「ここに登山道を引いた」と決め、AI が「その道に沿って案内板を整える」役割を引き受けている。

## 3. なぜ AI を使うのか — 単に楽をするためではない

教師が一人で全教材を書こうとすると、5 年で発見した縦軸を **学生に届けるための具体例を全部書き下ろす**作業に、また数年かかります。一方、AI を軸沿いの routing 装置として使うことで、教師は **発見した縦軸の保守と curation** に集中できます。

しかし、これには副作用があります。**AI が単独で生成すると、surface paraphrase（表面的な言い換え）に落ちることが多い**。同じ内容を別の語彙で繰り返すだけで、構造的に新しいものは生まれない。本講義ではこれを避けるために、**軸を image（structured corpus）として外部化** し、AI 生成が必ず軸沿いに routing されるよう設計しています。

詳細は同梱の `MOTIVATION.md` と `SPEC/` 配下を参照してください。

## 4. 設計の系譜 — O'Neill のハイフロンティアと Smalltalk

この教材の作り方は、最近の AI 流行で生まれた即席設計ではありません。**1970 年代の楽観的技術人文主義の系譜** の継承です。

### 4.1 O'Neill のハイフロンティア (1976) — オマージュの源

Princeton 大学の物理学者 Gerard K. O'Neill は、1976 年の著書 *The High Frontier: Human Colonies in Space* で、地球の限界を超えた人類の未来を提示しました。宇宙居住地、太陽光発電、Honest Sam という personal information assistant — 当時としては大胆な future vision を、**計算による定量的根拠** で裏付けた点が画期的でした。

> Gerard K. O'Neill, *The High Frontier: Human Colonies in Space*, William Morrow, 1976
> 続編: *2081: A Hopeful View of the Human Future*, Simon & Schuster, 1981

O'Neill の vision の核は三点:

1. **楽観 — ただし計算と実装で支える**。当時主流だった "Limits to Growth" 言説 (Club of Rome 1972) — 人類は資源枯渇で滅ぶ — に対し、大規模な技術的飛躍で乗り越えられるという反論を、**定量見積もりで** 示した
2. **協業の必要性**。宇宙居住地は一人や一国では作れない。**複数機関・複数世代の協力を最初から前提** に設計
3. **道具としての personal computer**。Dynabook と並ぶ時代の構想として、人間の認知能力を拡張する道具が中核

O'Neill 自身が **prototype を実際に組み**（mass driver の試作機）、**Space Studies Institute を設立** (1977)、講演で resources を集めて回った。**楽観は宣言で成立せず、prototype と数値と継続で成立する** — このプロジェクトの根本姿勢はここに由来します。

本講義の AI に対する姿勢は、O'Neill が宇宙開発に対して取った姿勢と同型です:

- AI の限界を「越えられない壁」とせず、**構造的応答** で延長を試みる（image による長期記憶の外部化）
- 楽観は宣言ではなく、**実装と数値** で支える（LanceDB / MCP / jupytext の実装、INV-G-009 として原則を invariant 化）
- 協業を前提とする — AI 単独でも人間単独でも到達できない知識生産を、**人間 × AI × image の三者の役割分担** で組み立てる

### 4.2 Smalltalk — O'Neill 的楽観の技術的実装

O'Neill が大きな vision を描いた同じ 1970 年代、Xerox PARC で **Alan Kay と Dan Ingalls が Smalltalk の image vision を実装** していました。O'Neill の "personal information assistant" の vision を、**動く実装** として地上に降ろした試みと読めます。

本講義はこの **二重のオマージュ** の上に立っています — **O'Neill の楽観・計算・協業** と、**Smalltalk の image による知識の持続性**。

- Alan Kay, *A Personal Computer for Children of All Ages* (1972) — image vision の起点
- Dan Ingalls, *Design Principles Behind Smalltalk* (BYTE, August 1981) — image 原則の最も簡潔な記述
- Adele Goldberg, *Smalltalk-80: The Interactive Programming Environment* (1984) — environment としての実装解説

### 4.3 Ingalls 1981 の三原則 — 本講義設計への直接対応

Ingalls (1981) は 17 の design principles を提示しましたが、本講義の設計に最も直接対応するのは次の三つです:

> **Personal Mastery**: *"If a system is to serve the creative spirit, it must be entirely comprehensible to a single individual."*
>
> （システムが創造的精神に仕えるためには、一人の個人にとって完全に理解可能でなければならない）

— 本講義の縦軸は、五年間 **一人の教師の頭の中に保持された** ものです。Personal Mastery の原則は、教師の理解可能性を超える複雑さを system に持ち込まない、という設計上の戒めとして機能します。

> **Uniform Metaphor**: *"A language should be designed around a powerful metaphor that can be uniformly applied in all areas."*
>
> （言語は、あらゆる領域に一貫して適用できる強力な metaphor を中心に設計されるべきである）

— 本講義における uniform metaphor は **「合成 (composition)」**です。construction（cum-struere）= synthesis（συν-θεσις）= composition（com-ponere）— 同じ語源を持つ概念を、作図・代数・コード・転生エッセイ・評価系のすべてに **均一に適用** しています。

> **Operating System**: *"An operating system is a collection of things that don't fit into a language. There shouldn't be one."*
>
> （OS とは、言語に収まらないものの寄せ集めである。OS は存在すべきでない）

— 本講義の image には、教材・対話・実装・履歴がすべて住んでいます。「教材は教材、講義ノートは別、AI 対話は別、コードは別」という分離が存在しない。image の外に置かれる「OS 的なもの」を最小化する設計です。

### 4.4 image の現代再演

Smalltalk の image とは「コード・データ・履歴・道具をひとつの問い合わせ可能な空間に住まわせる」設計で、半世紀前から技術として確立されていました。本講義の教材作成は、その vision を LanceDB（vector database）と MCP（Model Context Protocol）と Jupyter で組み直しただけです。**言語と道具は変わりましたが、image としての完結性・反射性・全体性はそのまま受け継いでいます**。

### 4.5 動いている Smalltalk を見る

文献だけでは抽象的に感じる方は、設計者 Dan Ingalls 本人の講義とデモを動画で確認できます:

1. **Object-Oriented Programming, lecture by Daniel Ingalls** — Smalltalk の中核開発者本人による OOP 講義。Smalltalk が単なる「クラスとオブジェクト」の言語ではなく、**統合された生きた system** として設計されたことが、開発者本人の言葉で語られる
   https://www.youtube.com/watch?v=Ao9W93OxQ7U

2. **Alto System Project: Dan Ingalls demonstrates Smalltalk** — Xerox Alto 上の Smalltalk の Ingalls 本人によるデモ。1970 年代の実装が、現代の GUI と OOP のほぼすべてを既に具備していたことが視覚的に確認できる
   https://www.youtube.com/watch?v=uknEhXyZgsg

3. **Smalltalk-78 simulator demo** — **Smalltalk が自分自身を書き換えるデモ**。実行中の image に対して、その image 自身のコードを修正し、修正が即座に動作に反映される様子が観察できる。これが Ingalls の言う **Reactive Principle**（すべてのコンポーネントは、観察と操作のために自身を意味ある形で提示できる）の実演であり、本講義の image vision の最も直接的な祖型です
   https://www.youtube.com/watch?v=eEz08IlcNMg

特に三本目は、**「システムが実行中に、自分自身を改変できる」** という設計が四十年前から実装として存在していた事実を、手触りで確認できる動画です。本講義で使う LanceDB image が、SPEC を ingest して MCP で自己検索できる（INV-G-008 自己参照ループ）構造の祖型がここにあります。

## 5. 全体構造の図

```
┌─────────────────────────────────────────────────────────┐
│  Higashida 先生の五年間の継続的学習・閃き                     │
│  (Babylonia → Menaechmus → Apollonius → Kepler            │
│   → Newton → Cauchy → Dandelin → Feynman → 五心対称性)     │
└──────────────────────┬──────────────────────────────────┘
                       │
                       │  軸として外部化
                       ▼
┌─────────────────────────────────────────────────────────┐
│  image core (Smalltalk へのオマージュ、半世紀前の知恵)         │
│  ┌───────────────────────────────────────────────────┐  │
│  │ 翻訳格子  (数⇄線分⇄代数⇄作図⇄code)                  │  │
│  │ 歴史軸    (4000 年を貫く恒等式)                       │  │
│  │ 段階構造  (手順→自己採点→表現多様性→symbolic⇄literal) │  │
│  │ 自問自答  (5 つのチェックリスト)                       │  │
│  └───────────────────────────────────────────────────┘  │
└──────────────────────┬──────────────────────────────────┘
                       │
                       │  軸に沿って routing (軸は発明しない)
                       ▼
┌─────────────────────────────────────────────────────────┐
│  AI (Claude 等) — 道具として                              │
│  軸沿いに教材を整形する。軸の外には踏み出さない                │
└──────────────────────┬──────────────────────────────────┘
                       │
                       │  生成 + 先生の curation
                       ▼
┌─────────────────────────────────────────────────────────┐
│  本教材 (あなたが今読んでいるもの)                          │
└─────────────────────────────────────────────────────────┘
```

## 6. 皆さんに保証できること

1. **教材の縦軸（系譜と中心命題）は AI の生成物ではない**。Higashida の五年間の発見と判断
2. **AI が言うことを丸呑みする必要はない**。AI は軸に沿った routing の一実装に過ぎず、別経路で同じ軸に到達することは歓迎される
3. **「これは AI が書いた感じがする」と感じた箇所は、おそらく軸を外れた surface paraphrase**。指摘してください。改善材料になります
4. **皆さんが AI と関わる場面（`%%ai` 等）はあくまで補助** で、評価は皆さんの手で組み立てた構成（construction）に対して行われます

## 6.5 段階的な移行 — AI の使い方は学期を通じて変わります（伏線）

本講義での AI の使い方は段階的に変わります:

- **最初** — 皆さんが他の授業で AI を使うときと同じく、単一の問いを投げて答えをもらう形から始めます
- **中盤** — 本講義専用の Jupyter AI (`@Codex` / GPT-5 mini) に移行
- **後半** — 「もし □□□期の ○○○ に転生したら」型の複数制約合成課題で、AI を **協業者** として使う形に進みます

段階が進むほど、皆さんが AI に **委託する部分** と **自分で curate する部分** の区別が明示的になります。最終的な課題では、AI との対話ログ・curation 過程・最終文・振り返りを併せて提出してもらうことで、AI と人間の役割分担そのものが評価対象になります。

これは §4.1 で示した O'Neill 的協業の精神の実装です。AI を「使う / 使わない」の二極ではなく、**「どう協業するか」** を学ぶ訓練の場として本講義を設計しています。詳細は各段階で改めて説明します。

## 6.7 教師の過剰防御 — 率直に書きます

ここまで本講義の設計を客観的に書いてきましたが、内面を率直に共有させてください。**Higashida は AI に対して過剰防御的な反応を持っています**。皆さんが AI への拒絶反応や違和感を持つかもしれないのと同じく、教える側にも防御反応がある。隠さずに書きます。

### 教師として、Higashida がやってきたこと

1. **AI Coding で授業ツールを開発した** — 4 ヶ月で約 5 万行。ggblab/GeoGebra.jl は AI 支援で短期間に立ち上げたもので、これ自体、伝統的な開発文化からは異例の規模と速度
2. **AI と改善点を議論し、自前の RAG-MCP を導入した**（lesson_04 から運用中、**今ここ**）— image vision を組み立て、SPEC を image に住まわせ、判断系譜を外部化する装置を作った。皆さんが今読んでいるこの教材も、その装置から生まれている
3. **学生が使える AI Agent を JupyterHub に準備し、皆さんに課題を課す** — `@Codex` (GPT-5 mini) を本講義の専用 AI として整えた。ただし、**何を合成したかのまとめは、皆さん自身の言葉で書いてほしい**

(3) の "ただし" がここでの過剰防御の核心です。「使っていい、でも自分の言葉で」と書いた瞬間、Higashida 自身が AI と人間の役割分担を不安に思っている表れでもあります。

### 学生側に予想される現状と、Higashida からの期待

1. **皆さんは色々な授業で、プロンプトエンジニアリングを教えられているはず** — 本講義はそれを学習対象にしません（プロジェクト不変原則 §0.2）。プロンプト工夫より、**合成の質** を評価します
2. **「Wikipedia を疑え」と教えられている** と思いますが、注意点があります。本講義で参照する最新の知見には、**英語版 Wikipedia に更新済みで日本語版にはまだ反映されていない** 情報が含まれます（たとえば近年地中海で発見された写本群に関する記述など）。「Wikipedia は信用できない」は一般則として正しいが、**英語版で cutting-edge 情報の所在を確認できる場合がある** ことを覚えておいてください
3. **「死海文書」と聞いて Evangelion 等のサブカル連想を持つかもしれませんが**、本来の歴史的意義は別です。1947〜1956 年に発見されたヘブライ語写本群は、紀元前 3 世紀〜紀元 1 世紀の旧約聖書テキストを含み、紀元 1000 年頃の Masoretic Text とほぼ一致する。**旧約聖書が千年以上ほぼ原型を保っていた** ことの textual criticism としての証拠です。

   **ただし — 日本にも同型の対象があります**。古事記（712 CE）と日本書紀（720 CE）は、大陸伝承から **ほぼ独立に編纂された** 創世神話を、二書として 8 年差で並走させた文書群です。両書は内容も神統も substantially 異なり（source criticism の素材）、各々に複数の写本系統が存在（古事記の真福寺本・兼永筆本など）。死海文書 vs Masoretic と完全に同型の textual criticism の対象になり得る資産です。

   ところが、日本では古事記/日本書紀を死海文書と同じ厳密性で扱う議論が pop culture にも academia にも広く浸透していません。**「外の textual 遺物は exotic にオカルト扱い、自国の同型物は検証の対象外に置く」非対称** が、私たちの文化に染み付いている。本講義で育てたい「軸の検証感覚」を学生に渡す際、この非対称が見えない障壁になります。

   本講義の image vision（軸を持続させて伝える設計）の遠い祖型として、**死海文書 + 古事記/日本書紀の双方** が等しく参照されるべき、というのが Higashida の立場です
4. **皆さんと AI が協業する O'Neill 的楽観的未来の一像を、本講義で示したい** — ただし、これは宣言で成立しません。§4.1 の O'Neill が prototype と数値と継続で支えたように、**皆さん自身も働き者である必要があります**。AI に委託しても、curation の労働と自分の言葉での総括は、皆さんの手で

最後の点が、本講義からの最も真剣な期待です。AI 委託は許容、curation と総括は皆さんの仕事。これが O'Neill 的協業の、皆さんに向けた具体的な形です。

## 7. 皆さんへの問いかけ

本講義の縦軸を理解した上で、AI を **どこまで使い、どこで使わないか** を、皆さん自身が選び取る訓練の場として、この教材を使ってください。AI を「拒絶」も「全面信頼」もせず、**道具として何を担わせ、何を自分で担うか** を判断できるようになることが、社会に出る前に身につけるべきリテラシです（[2.5 節](../conversations/2026-05-12/SUMMARY_chatlogs_for_textbook.md) も参照）。

第 4 回以降の本論で、その判断の具体例を一つずつ体験していきます。
